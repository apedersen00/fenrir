```python
import json
import matplotlib.pyplot as plt

# Load training statistics
with open("training_stats.json", "r") as f:
    stats = json.load(f)

epochs = range(1, len(stats["train_loss"]) + 1)

plt.figure(figsize=(16, 10))

plt.subplot(2, 3, 1)
plt.plot(epochs, stats["train_loss"], marker='o', label="Train Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Train Loss")
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(epochs, stats["train_acc"], marker='o', label="Train Acc", color="green")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Train Accuracy")
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(epochs, stats["test_acc"], marker='o', label="Test Acc", color="orange")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Test Accuracy")
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(epochs, stats["train_batch_time"], marker='o', label="Avg Batch Time", color="purple")
plt.xlabel("Epoch")
plt.ylabel("Seconds")
plt.title("Avg Batch Time per Epoch")
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(epochs, stats["train_epoch_time"], marker='o', label="Epoch Time", color="red")
plt.xlabel("Epoch")
plt.ylabel("Seconds")
plt.title("Total Epoch Time")
plt.grid(True)

plt.tight_layout()
plt.show()
```
